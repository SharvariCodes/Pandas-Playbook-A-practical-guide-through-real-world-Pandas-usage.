## **14. How many people have a credit card that expires in 2020?**

```python
def expire_year():
    count = 0
    for date in data['CC Exp Date']:
        if date.split('/')[1] == '20':
            count = count + 1
    return count
```

```python
expire_year()
```

```python
data[data['CC Exp Date'].apply(lambda x:x[3:] == '20')].count()
```
